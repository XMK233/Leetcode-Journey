## The difference is generated by using filecmp. 
* It is a very easy and fast way. 

```python
import os, filecmp

PATH = r"J:\ModelStoreData\AIHub\2019-06-04\TFModule"

module_list = os.listdir(PATH)

def diff_files_in_different_dirs(dir1, dir2):
    walk_a = os.walk(dir1)
    walk_b = os.walk(dir2)
    list_a = []
    list_b = []
    for (dirpath, dirnames, filenames), (dp, dns, fns) in zip(walk_a, walk_b):
        for filename in filenames:
            list_a.append(os.path.join(dirpath, filename))
        for fn in fns:
            list_b.append(os.path.join(dp, fn))
    if len(list_a) != len(list_b):
        print("some file is missing. ")
        return False
    else:
        cmp_rst = []
        for a, b in zip(list_a, list_b):
            if "descriptor.txt" in a:
                continue
            if not filecmp.cmp(a, b, shallow=False):
                print(a)
                print(b)
                print("they are different. ")
                return False
#                 cmp_rst.append(False)
#         if False in cmp_rst:
#             return False
    return True

def test_method():
    module_count = 1
    modules_without_multiple_versions = []
    for module in module_list:
        module_path = os.path.join(PATH, module)
        if not os.path.isdir(module_path):
            continue
        file_dir = os.path.join(module_path, "file")
        versions_of_checkpoint = os.listdir(file_dir)
        versions_of_checkpoint.sort()
        if len(versions_of_checkpoint) <= 1:
            modules_without_multiple_versions.append(module)
            continue
        print("{}. {} #versions: {}".format(module_count, module, len(versions_of_checkpoint)))
        module_count += 1
        for index in range(len(versions_of_checkpoint) - 1):
            diff_files_in_different_dirs(
                os.path.join(file_dir, versions_of_checkpoint[index]),
                os.path.join(file_dir, versions_of_checkpoint[index + 1])
            )
        print("\n")
        
    print("number of modules without multiple versions: {}".format(len(modules_without_multiple_versions)))

test_method()
```
