```python
import os, json

JSON_FILE = r"J:\ModelStoreData\AIHub\2019-06-04\TFModule_Info.json"

def find_how_many_models_has_multiple_versions():
    with open(JSON_FILE, "r") as jf:
        items = json.load(jf)

    modules = {}
    multi_version_modules = {}
    
    for item in items:
        name = item["name"]
        version = item["version"]
        if name not in modules:
            modules[name] = [version]
        else:
            modules[name].append(version)

    for module in modules:
        if len(modules[module]) >= 2:
#             print(module, modules[module])
            multi_version_modules[module] = modules[module]
    
    return multi_version_modules, modules
```

```python
 multi_version_modules, modules = find_how_many_models_has_multiple_versions()
```

```python
def sort_and_print_dict(d, f):
    with open(f, "w") as fw:
        fw.write("module,versions\n")
        l = list(d.keys())
        print(len(l))
        l.sort()
        for i in l:
            fw.write("{},{}\n".format(i, len(d[i])))
            print(i, len(d[i]))
sort_and_print_dict(modules, "product_versions.csv")
```
