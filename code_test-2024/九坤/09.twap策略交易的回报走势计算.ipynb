* 将买入和卖出的两类分开。分两个股票集合: stock_buy, stock_sell. 
* 建两个表：
    * 一个表t1表示，每一笔交易的量，比如每一笔交易100笔。
    * 另一个表t2表示，到当前时刻，总共交易的笔数，比如100，200，300.....
* 对于 stock_buy：
    * t2\[stock_buy\] * udata\[stock_buy\] = df_buy_theoretical
    * t1\[stock_buy\] * udata\[stock_buy\] = df_buy_every
    * df_buy_every按行计算累积和，得到新的表df_buy_actually. 
    * df_buy_earn = df_buy_theoretical - df_buy_every
* 对于 stock_sell: 
    * t2\[stock_sell\] * udata\[stock_sell\] = df_sell_theoretical
    * t1\[stock_sell\] * udata\[stock_sell\] = df_sell_every
    * df_sell_every按行计算累积和，得到新的表df_sell_actually.
    * df_sell_earn = df_sell_actually - df_sell_every
* 总共合并起来 df_earn = pd.concat(\[df_buy_earn, df_sell_earn\], index=1)
* df_earn.sum(axis=1) ## 总之就是按行加和，就能得到每一个时刻的收益情况了。

```python
## 怎么选择 stock_buy 呢？
ope_num = udata.shape[0]-1 ## 交易的次数。
alpha_T = alpha.T
alpha_T["tar_minus_init"] = alpha_T.TargetPos - alpha_T.InitPos
alpha_T["total_quant"] = alpha_T["tar_minus_init"].abs()
alpha_T["quant_per_ope"] = alpha_T["total_quant"]/ope_num
alpha_T
```

```python
stock_sell = list(alpha_T[alpha_T.tar_minus_init <= 0].index)
stock_buy = list(alpha_T[alpha_T.tar_minus_init > 0].index)
```

```python
t1 = pd.concat([alpha_T[["quant_per_ope"]].T] * ope_num).reset_index(drop=True)
t2 = t1.cumsum()
```

```python
t1
```

```python
t2
```

```python
df_buy_theoretical = t2[stock_buy] * udata[stock_buy]
df_buy_every = t1[stock_buy] * udata[stock_buy]
df_buy_actually = df_buy_every.cumsum()
df_buy_earn = df_buy_theoretical - df_buy_actually
df_buy_earn
```

* 对于 stock_sell: 
    * t2\[stock_sell\] * udata\[stock_sell\] = df_sell_theoretical
    * t1\[stock_sell\] * udata\[stock_sell\] = df_sell_every
    * df_sell_every按行计算累积和，得到新的表df_sell_actually.
    * df_sell_earn = df_sell_actually - df_sell_every

```python
df_sell_theoretical = t2[stock_sell] * udata[stock_sell]
df_sell_every = t1[stock_sell] * udata[stock_sell]
df_sell_actually = df_sell_every.cumsum()
df_sell_earn = df_sell_actually - df_sell_every
df_sell_earn
```

```python
df_earn = pd.concat([df_buy_earn, df_sell_earn], axis=1)
df_earn
```
