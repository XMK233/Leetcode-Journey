### Fine-tuning: change default batch norm momentum to 0.99 and make it configurable.
* Added signature:  
    * name: `image_classification_with_bn_hparams` and `image_feature_vector_with_bn_hparams`:
        * inputs: `batch_norm_momentum`

```python
import hashlib

def see_in_changelog_1():
    modules = Changelog_module["Fine-tuning: change default batch norm momentum to 0.99 and make it configurable."]
    for module, api_diff_result in modules.items():
        if "classification" in module:        
            s = str(api_diff_result["dictionary_item_added"]["root['image_classification_with_bn_hparams']"]["inputs"]["batch_norm_momentum"]) + str(api_diff_result["dictionary_item_added"]["root['image_feature_vector_with_bn_hparams']"]["inputs"]["batch_norm_momentum"])
            print(
                hashlib.md5(s.encode('utf-8')).hexdigest(), module
            )
            print(s)
    
    return

def see_in_changelog_2():
    modules = Changelog_module["Fine-tuning: change default batch norm momentum to 0.99 and make it configurable."]
    for module, api_diff_result in modules.items():
        if "feature_vector" in module:        
            s = str(api_diff_result["dictionary_item_added"]["root['image_feature_vector_with_bn_hparams']"]["inputs"]["batch_norm_momentum"])
            print(
                hashlib.md5(s.encode('utf-8')).hexdigest(), module
            )
            print(s)
    
    return

see_in_changelog_1()
see_in_changelog_2()
        
```
