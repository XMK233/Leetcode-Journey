* And based on this situation, merging is based on log_module.json

```python
import json

def read_json(file_path):
    with open(file_path, "r") as fp:
        dic = json.load(fp)
    return dic

def write_json(file_path, content, indent = 4, sortKey = True):
    with open(file_path, "w") as fp:
        json.dump(content, fp, indent=indent, sort_keys=sortKey)
    return

log_module = read_json("log_module_v1.json") #read_json("log_module.json")
api_diffing = read_json("APIDiff_changelog.json")

for log in log_module:
    modules = log_module[log]
    for module_version in modules:
        module = module_version.split("@")[0]
        version = module_version.split("@")[1].split()[1]
        diff_result = api_diffing.get(module)
        if diff_result == None:
            continue
        for key, value in diff_result.items():
            print(key)
            if key.endswith(version):
                modules[module_version] = value
                # module_version.append(value)

write_json("Merged_ChangelogGrouping_APIDiffing_v1.json", log_module)

# for log in log_module:
#     modules = log_module[log]
#     for module_version in modules:
#         module = module_version[0]
#         version = module_version[1].split()[1]
#         diff_result = api_diffing.get(module)
#         if diff_result == None:
#             continue
#         for key, value in diff_result.items():
#             if key.endswith(version):
#                 module_version.append(value)

# write_json("Merged_ChangelogGrouping_APIDiffing.json", log_module)
```
